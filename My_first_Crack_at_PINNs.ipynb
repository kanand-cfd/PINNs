# Architecture of a Physics Informed Neural Network

The idea of neural networks as universal approximators is widely accepted now. Physics Informed Neural Networks (or PINNs) leverage domain knowledge to overcome the issue of imbalanced data & data shortage commonly encountered for a generic neural network. Let us begin with a simple ordinary differential equation (ODE) with some measured data as the initial and boundary conditions. 

The idea is to solve a differential equation by representing the solution through a neural network and thereby training the network so that it satisfies the conditions imposed by the differential equation.

Consider a system of ODEs

$$u' = f(u,t) $$

Let us approximate the solution using a neural network

$$u(t) \equiv \mathcal{N}_N (t) $$

If $\mathcal{N}_N (t)$ was the true solution, then it would hold true for $\mathcal{N'}_N(t) = f(\mathcal{N}_N(t), t)$ for all $t$. Thus, the associated loss function can be written as

$$\mathcal{L}(\omega) = \Sigma_i \left( \frac{d \mathcal{N}_N(t_i)}{dt} - f(\mathcal{N}_N(t_i), t_i) \right)^2 $$


When this loss function is minimized, the neural network approximately solves the differential equation.

In order to handle the initial conditions, they are introduced in the loss function as

$$\mathcal{L}(\omega) = \Sigma_i \left( \frac{d \mathcal{N}_N(t_i)}{dt} - f(\mathcal{N}_N(t_i), t_i) \right)^2 + \left( \mathcal{N}_N - u(0) \right)^2$$

$\omega$ here represents the parameters that define the neural network. Hence the problem now reduces down to computing the weights of the neural network for minimizing the loss function.

### Simple ODE

$$\frac{d u}{d t} = \cos{\pi t}$$

Initial Condition : 

$$u(0) = 1$$

The exact solution is given as:

$$u(t) = \frac{1}{2\pi} \sin{2 \pi t} + 1$$

The implementation is given below:

### Import the relevant libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import random
```

### Construct the Neural Netwerk $\mathcal{N}_N (t)$ and define the hyper-parameters 

```python
NN = tf.keras.models.Sequential([
    tf.keras.layers.Input(1,),
    tf.keras.layers.Dense(units=32, activation='tanh'),
    tf.keras.layers.Dense(units=1)
])

NN.summary()
```

```python
optm = tf.keras.optimizers.legacy.Adam(learning_rate=0.001)
```

### Define ODE system and loss function

* ODE loss:

$$\mathcal{L}_{ODE} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{d \mathcal{N}_N(t_i)}{d t} - \cos{2\pi t_i} \right)^2$$

* I.C. loss:

$$\mathcal{L}_{IC} = \frac{1}{n} \sum_{i=1}^{n} \left( \mathcal{N}_N(0) - 1 \right)^2$$

* Total loss:

$$\mathcal{L}_{total} = \mathcal{L}_{ODE} + \mathcal{L}_{IC} $$

```python
def ode_system(t, net):
    t = t.reshape(-1, 1)
    t = tf.constant(t, dtype=tf.float32)
    t0 = tf.zeros((1,1))
    one = tf.ones((1,1))
    
    with tf.GradientTape() as tape:
        tape.watch(t)
        
        u = net(t)
        u_t = tape.gradient(u,t)

    ode_loss = u_t - tf.math.cos(2.0 * np.pi * t)
    IC_loss = net(t0) - one
    
    squared_loss = tf.square(ode_loss) + tf.square(IC_loss)
    total_loss = tf.reduce_mean(squared_loss)
    
    return total_loss
```

### Training & Prediction

```python
train_t = (np.random.rand(30)*2).reshape(-1, 1)
train_loss_record = []

for itr in range(10000):
    with tf.GradientTape() as tape:
        train_loss = ode_system(train_t, NN)
        train_loss_record.append(train_loss)
        
        grad_w = tape.gradient(train_loss, NN.trainable_variables)
        optm.apply_gradients(zip(grad_w, NN.trainable_variables))
    
    
    if itr % 1000 == 0:
        print(train_loss.numpy())

plt.figure(figsize=(10,8))
plt.plot(train_loss_record)
plt.show()
```

```python
test_t = np.linspace(0, 2, 100)

train_u = np.sin(2.0*np.pi*train_t)/(2.0*np.pi) + 1.0
true_u = np.sin(2.0*np.pi*test_t)/(2.0*np.pi) + 1.0
pred_u = NN.predict(test_t).ravel()

plt.figure(figsize = (10,8))
plt.plot(train_t, train_u, 'ok', label = 'Train')
plt.plot(test_t, true_u, '-k',label = 'True')
plt.plot(test_t, pred_u, '--r', label = 'Prediction')
plt.legend(fontsize = 15)
plt.xlabel('t', fontsize = 15)
plt.ylabel('u', fontsize = 15)
plt.show()
```
